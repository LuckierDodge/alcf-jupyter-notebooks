```python
# Grab an array of all the files
fns = glob.glob('data/darshan/' + str(year) + '/' + str(month) + '/' + str(day) + '/*.darshan')
# fns
```

```python
def extract_file_data(filename):
    print("Extracting from", filename, "...")
    try:
        filehandle = os.popen('/home/rlewis/dev/darshan/bin/darshan-parser --all ' + filename)#.read()
    except:
        print("error when run darshan-parser for %s" % filename)
        return None

    jid, start_time, end_time, nprocs = None, None, None, None
    for line in filehandle:
        if line.startswith('# jobid:'): 
            jid = int(line.split()[-1])
            continue

        if line.startswith('# start_time:'): 
            start_time = int(line.split()[-1])
            continue

        if line.startswith('# end_time:'): 
            end_time = int(line.split()[-1])
            continue

        if line.startswith('# nprocs:'): 
            nprocs = int(line.split()[-1])
            continue

        if line.startswith('#<rank>'):
            break

    file_rows = []
    for line in filehandle:
        if line.startswith('total_CP_'): break
        file_rows.append(line)
    fdf = pd.DataFrame([x.split() + [jid, start_time, end_time, nprocs] for x in file_rows],
                       columns=('rank', 'filename', 'counter', 'value', 'fn_suffix', 'mount_pt', 'fs_type', 'jid', 'start_time', 'end_time', 'nprocs'))
    print("Extraction for", filename, "done, writing to csv")
    fdf.to_csv('data/darshan_csv/' + str(year) + str(month) + str(day) + '/' + str(os.path.basename(filename)) + '.csv')
    print(os.path.basename(filename), " done!")
#     return fdf
```

```python
def create_csv(filenames):
    df = pd.DataFrame()
    print("Extracting files for ", year, month, day, "...")
    if __name__ == '__main__':
        with Pool(128) as p:
            p.map(extract_file_data, filenames)
            p.close()
#     extract_file_data(filenames[0])
    print("All files extracted, starting append")
    for filename in filenames:
        tempdf = pd.read_csv('data/darshan_csv/' + str(year) + str(month) + str(day) + '/' + str(os.path.basename(filename)) + '.csv')
        df = df.append(tempdf)
#     for fdf in fdfs:
#         df = df.append(fdf)
#     print("Finished append, writing to csv...")
#     df.to_csv('data/darshan_csv/darshan_%s_%s_%s.csv', year, month, day)
    print("Csv written.")
```

```python
create_csv(fns)
```
