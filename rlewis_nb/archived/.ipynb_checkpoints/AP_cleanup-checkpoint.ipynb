```python
import pandas as pd
import numpy as np
import glob
from matplotlib import pyplot as plt
from sklearn import metrics
pd.options.display.max_columns = 1200
pd.options.display.max_rows = 1200
```

# Load the Data

```python
# df = pd.read_csv('data/autoperf-2019-01.csv', header=0).sample(frac=1)

li = []
for filename in glob.glob("/home/luckierdodge/repos/jupyter-notebooks/data/autoperf_complete/autoperf-201*.csv"):
    frame = pd.read_csv(filename, header=0)
    li.append(frame)
df = pd.concat(li, axis=0, ignore_index=True)

df.describe()
```

```python
# Drop NANs
df.dropna(inplace=True);
df.describe()
```

# Filter by Core Hours

```python
# counts = df.zero_execName.value_counts().to_dict()
# df_filtered = pd.DataFrame()
# for name in counts:
#     if counts[name] >= 500:
#         df_filtered = df_filtered.append(df[df["zero_execName"] == name])
# df = df_filtered.copy()

names = df.zero_execName.unique()
df_filtered = pd.DataFrame()
for name in names:
    df_name = df[df['zero_execName'] == name].copy()
    corehours = df_name.zero_numRanks * df_name.zero_elapsedTime
    df_name['coreHours'] = corehours
    if corehours.sum() >= 60 * 60 * 1000000:
        df_filtered = df_filtered.append(df_name)
```

```python
df_filtered.describe()
```

```python
df_filtered.to_csv(path_or_buf='/home/luckierdodge/repos/jupyter-notebooks/data/autoperf_filtered_1000000ch', index=False)
```
