```python
import pandas as pd
import numpy as np
import tensorflow as tf
pd.options.display.max_columns = 300
```

```python
data_2015 = pd.read_csv('./data/ANL-ALCF-AUTOPERF-MIRA_20150101_20151231.csv', header=0, sep=',', quotechar='"')
data_2016 = pd.read_csv('./data/ANL-ALCF-AUTOPERF-MIRA_20160101_20161231.csv', header=0, sep=',', quotechar='"')
data_2017 = pd.read_csv('./data/ANL-ALCF-AUTOPERF-MIRA_20170101_20171231.csv', header=0, sep=',', quotechar='"')
data_2018 = pd.read_csv('./data/ANL-ALCF-AUTOPERF-MIRA_20180101_20181231.csv', header=0, sep=',', quotechar='"')
data_2019 = pd.read_csv('./data/ANL-ALCF-AUTOPERF-MIRA_20190101_20190430.csv', header=0, sep=',', quotechar='"')
```

```python
# Filter the data here
filter_hours = 10
for data in [data_2015, data_2016, data_2017, data_2018, data_2019]:
    print(data.shape[0])
    data.drop(data[data.PROC_ELAPSEDTIME < filter_hours].index, inplace = True)
    print(data.shape[0])
```

### Key

PROC_CSJOBID - Task IDs

PROC_BATCHJOBID - Job IDs

PROC_ELAPSEDTIME - Job Total Time

```python
averageDict = {'year': ['2015', '2016', '2017', '2018', '2019'],
              'job_id_count': [
                  data_2015.PROC_BATCHJOBID.unique().size,
                  data_2016.PROC_BATCHJOBID.unique().size,
                  data_2017.PROC_BATCHJOBID.unique().size,
                  data_2018.PROC_BATCHJOBID.unique().size,
                  data_2019.PROC_BATCHJOBID.unique().size,
              ],
              'task_id_count': [
                  data_2015.PROC_CSJOBID.unique().size,
                  data_2016.PROC_CSJOBID.unique().size,
                  data_2017.PROC_CSJOBID.unique().size,
                  data_2018.PROC_CSJOBID.unique().size,
                  data_2019.PROC_CSJOBID.unique().size
              ]}
averageData = pd.DataFrame(data=averageDict)
averageData['tasks_per_job'] = averageData['task_id_count'] / averageData['job_id_count']
print(averageData)
```

```python
%matplotlib inline
averageData.plot(x="year", 
                 y="tasks_per_job", 
                 kind='bar', 
                 title="Average Number of Tasks Per Job vs. Year for Autoperf Data on Mira\n2015 to 2019", 
                 figsize=(10,10), 
                 color='orange', 
                 legend=False)
```

```python
years = [data_2015, data_2016, data_2017, data_2018, data_2019]
yearstrings = ['2015', '2016', '2017', '2018', '2019']
y = {'year': yearstrings, 'Traditional HPC': np.zeros(5, dtype=float), 'Workflow': np.zeros(5, dtype=float)}
job_task_data = pd.DataFrame(data=y)

for i in range(0, 5):
    one = 0.0
    many = 0.0
    for job in years[i].PROC_BATCHJOBID.unique():
        tasks = years[i][years[i]['PROC_BATCHJOBID'] == job].PROC_CSJOBID.unique().size
        if tasks == 1:
            one += 1.0
        else:
            many += 1.0
    total = one + many
    job_task_data.iat[i, 1] = one / total
    job_task_data.iat[i, 2] = many / total
job_task_data.plot(kind='bar', x='year', y=['Traditional HPC', 'Workflow'], figsize=(10,10), stacked=True)
print(job_task_data)
```

```python
years = [data_2015, data_2016, data_2017, data_2018, data_2019]
yearstrings = ['2015', '2016', '2017', '2018', '2019']
y = {'year': yearstrings, 'Traditional HPC': np.zeros(5, dtype=float), 'Workflow': np.zeros(5, dtype=float)}
job_task_data = pd.DataFrame(data=y)

for i in range(0, 5):
    one = 0.0
    many = 0.0
    for job in years[i].PROC_BATCHJOBID.unique():
        tasks = years[i][years[i]['PROC_BATCHJOBID'] == job].PROC_CSJOBID.unique().size
        if tasks == 1:
            one += years[i][years[i].PROC_BATCHJOBID == job].iloc[0].PROC_ELAPSEDTIME
        else:
            many += years[i][years[i].PROC_BATCHJOBID == job].iloc[0].PROC_ELAPSEDTIME
    total = one + many
    job_task_data.iat[i, 1] = one / total
    job_task_data.iat[i, 2] = many / total
job_task_data.plot(kind='bar', x='year', y=['Traditional HPC', 'Workflow'], figsize=(10,10), stacked=True)
print(job_task_data)
```

```python
years = [data_2015, data_2016, data_2017, data_2018, data_2019]
yearstrings = ['2015', '2016', '2017', '2018', '2019 Extrapolated']
y = {'year': yearstrings, 'Traditional HPC': np.zeros(5, dtype=int), 'Workflow': np.zeros(5, dtype=int)}
job_task_data = pd.DataFrame(data=y)

for i in range(0, 5):
    one = 0
    many = 0
    for job in years[i].PROC_BATCHJOBID.unique():
        tasks = years[i][years[i]['PROC_BATCHJOBID'] == job].PROC_CSJOBID.unique().size
        if tasks == 1:
            one += years[i][years[i].PROC_BATCHJOBID == job].iloc[0].PROC_ELAPSEDTIME
        else:
            many += years[i][years[i].PROC_BATCHJOBID == job].iloc[0].PROC_ELAPSEDTIME
    if (i < 4):
        job_task_data.iat[i, 1] = one
    else:
        job_task_data.iat[i, 1] = one * 3
    if (i < 4):
        job_task_data.iat[i, 2] = many
    else:
        job_task_data.iat[i,2] = many * 3
print(job_task_data)
job_task_data.plot(kind='bar', x='year', y=['Traditional HPC', 'Workflow'], figsize=(10,10), stacked=True)
```
