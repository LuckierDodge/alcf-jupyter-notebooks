# Resource.Trace1 Creation

```python
djc_df = djc_df[~djc_df['QUEUE_NAME'].str.startswith('R.')]
```

```python
joiner = ":1]["
queue_list = "[" + joiner.join(djc_df.QUEUE_NAME.unique()) + "]"
print(queue_list)
```

```python
rdf = pd.DataFrame({'Resource Type': ['COMPUTENODE'],
                'Event Type': ['AVAILABLE'],
                'Event Time': [0],
                'Resource ID': ["MIR%04x" % 0],
                'Resource Manager Name': ['PBS1'],
                'Configured Swap': [0],
                'Configured Memory': [16384],
                'Configured Disk': [100000],
                'Configured Processors': [16],
                'Resource Frame Location': [1],
                'Resource Slot Location': [1],
                'Resource Slot Use Count': [1],
                'Node Operating System': ['LINUX'],
                'Node Architecture': ['BLUEGENEQ'],
                'Configured Node Features': ['[NONE]'],
                'Configured Run Classes': [queue_list],
                'Configured Network Adapters': ['[NONE]'],
                'Relative Resource Speed': [1],
                'RESERVED FIELD 1': ['[NONE]'],
                'RESERVED FIELD 2': ['[NONE]'],
                'RESERVED FIELD 3': ['[NONE]'],
               })
for x in range(1, 49152):
    rdf = rdf.append({'Resource Type': 'COMPUTENODE',
                'Event Type': 'AVAILABLE',
                'Event Time': 0,
                'Resource ID': "MIR%04x" % x,
                'Resource Manager Name': 'PBS1',
                'Configured Swap': 0,
                'Configured Memory': 16384,
                'Configured Disk': 100000,
                'Configured Processors': 16,
                'Resource Frame Location': 1,
                'Resource Slot Location': 1,
                'Resource Slot Use Count': 1,
                'Node Operating System': 'LINUX',
                'Node Architecture': 'BLUEGENEQ',
                'Configured Node Features': '[NONE]',
                'Configured Run Classes': queue_list,
                'Configured Network Adapters': '[NONE]',
                'Relative Resource Speed': 1,
                'RESERVED FIELD 1': '[NONE]',
                'RESERVED FIELD 2': '[NONE]',
                'RESERVED FIELD 3': '[NONE]',
               }, ignore_index=True)
rdf.head()
```

```python
rdf.to_csv('../data/Resource.TraceMira1', sep=' ', header=False, index=False)
```

# Workload.Trace Creation

```python
wdf = pd.DataFrame()
wdf['JobID'] = djc_df['COBALT_JOBID']
wdf['Nodes Requested'] = djc_df['NODES_REQUESTED'].astype(int)
wdf['Tasks Requested'] = djc_df['CORES_REQUESTED'].astype(int)
wdf['User Name'] = djc_df['USERNAME']
wdf['Group Name'] = djc_df['PROJECT_NAME']
wdf['Wallclock Limit'] = djc_df['WALLTIME_SECONDS'].astype(int)
wdf['Job Completion State'] = 'Completed' # Ignoring exit status for simplicity's sake, shouldn't affect our results
wdf['Required Class'] = '[' + djc_df['QUEUE_NAME'] + ':1]'
wdf['Submission Time'] = (pd.to_datetime(djc_df['QUEUED_TIMESTAMP']).astype(np.int64) / 1000000000).astype(int)
wdf['Dispatch Time'] = (pd.to_datetime(djc_df['START_TIMESTAMP']).astype(np.int64) / 1000000000).astype(int)
wdf['Start Time'] = (pd.to_datetime(djc_df['START_TIMESTAMP']).astype(np.int64) / 1000000000).astype(int)
wdf['Completion Time'] = (pd.to_datetime(djc_df['END_TIMESTAMP']).astype(np.int64) / 1000000000).astype(int)
wdf['Required Network Adapter'] = "[NONE]"
wdf['Required Node Architecture'] = "[NONE]"
wdf['Required Node Operating System'] = "[NONE]"
wdf['Required Node Memory Comparison'] = ">="
wdf['Required Node Memory'] = 1
wdf['Required Node Disk Comparison'] = ">="
wdf['Required Node Disk'] = 1
wdf['Required Node Attributes'] = "[NONE]"
wdf['System Queue Time'] = wdf['Submission Time'] # This may or may not be right
wdf['Tasks Allocated'] = djc_df['CORES_USED'].astype(int)
wdf['Required Tasks Per Node'] = -1
wdf['QOS'] = "[NONE]"
wdf['JobFlags'] = "[NONE]" # May need revision
wdf['Account Name'] = djc_df['PROJECT_NAME']
wdf['Executable'] = djc_df['COMMAND']
wdf['Comment'] = "[NONE]"
wdf['Bypass Count'] = -1
wdf['ProcSeconds Utilized'] = djc_df['USED_CORE_HOURS'] * 3600
wdf['Partition Name'] = '[NONE]'
wdf['Dedicated Processors per Task'] = 1
wdf['Dedicated Memory per Task'] = '[NONE]'
wdf['Dedicated Disk per Task'] = '[NONE]'
wdf['Dedicated Swap per Task'] = '[NONE]'
wdf['Start Date'] = 0
wdf['End Date'] = 0
wdf['Allocated Host List'] = '[NONE]'
wdf['Resource Manager Name'] = '[NONE]'
wdf['Required Host Mask'] = '[NONE]'
wdf['Reservation'] = '[NONE]'
wdf['Set Description'] = '[NONE]'
wdf['Application Simulator Data'] = '[NONE]'
wdf['RESERVED FIELD 1'] = '[NONE]'
```

```python
wdf.head()
```

```python
wdf.to_csv('../data/Workload.TraceMira1', sep=' ', header=False, index=False)
```
